```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
##Pawe≈Ç Radzik zad1
def f(x):
    return x*x+5;
```

```python
x=np.linspace(-1,1)

plt.plot(x,f(x))
plt.legend("x^2+5")
plt.title("<-1,1>")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
x=np.linspace(-6,6)
plt.plot(x,f(x))
plt.legend("x^2+5")
plt.title("<-6,6>")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
x=np.linspace(0,5)
plt.plot(x,f(x))
plt.legend("x^2+5")
plt.title("<0,5>")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
